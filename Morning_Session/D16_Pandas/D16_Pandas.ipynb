## Numpy
* Numpy : Numerical python
    * **Advantages:**
        * convenience and speed 
            * Numpy is much faster than the standard python ways to do computations
            * Can write vectorised code on numpy arrays, not on lists
            * which is convenient to read and write
            * It takes Less memory as compared to python lists 
            * Can handle the vast amount of data 

    a. Used for scientific computing and data analysis
    b. Written as ndarray (n-dimensional , homogeneous) or array
        * Commonly integers or floats
    c. Creating Numpy Arrays:
        * 1D, 2D, 3D array
        * Convert lists, tuple to np.arrays

    d. Initialize Array:

        * np.ones(shape,dtype)
        * np.zeros(shape,dtype)
        * np.random.random(shape)
        * np.arange(start,end,step)
        * np.linspace(start,end,length of array)
        * np.full(shape,value)
        * np.tile(array,number of repetitions) or (array,shape)
        * np.eye(value,dtype) - identical matrix
        * np.random.randint(start,end,shape)
        * np.nonzero(array)
        * max,min,median,mode
        * array Transpose
        * np.vstack(array)
        * np.hstack(array
        * argsort,argmin,argmax,np.log10,pi
        
    e. Inspect the structure and content of array

        * np.shape(array)
        * np.dtype()
        * np.ndim()
        * np.itemsize()
        * np.reshape(factors of size of array)

## Pandas Dataframe
* Dataframe is the most widely used data-structure in data analysis. It is a table with rows and columns, with rows having an index and columns having meaningful names.

* Creating dataframes from dictionaries

* There are various ways of creating dataframes, such as creating them from dictionaries, JSON objects, reading from txt, CSV     files, etc.

*One of the main purposes of pandas is to deal with tabular data, i.e. data that comes from tables or spreadsheets.*

````python
# creating DF using dict
import pandas as pd
di = {   "name": ["chandu",'shaik',"keerthi","srikanthi"],
         "age" : [12,34,22,34]
     }
df = pd.DataFrame(di)
df
````

````python
import pandas as pd
di = {   "name": pd.Series(["chandu",'shaik',"keerthi","srikanthi"],index = ['a','b','c','d']),
         "age" : pd.Series([12,34,22,34],index = ['a','b','c','d'])
     }
df = pd.DataFrame(di)
df
````
